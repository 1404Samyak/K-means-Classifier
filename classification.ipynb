```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV,StratifiedKFold
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('pollution.csv')
data
```

```python
data.isna().sum()
```

```python
X=data.drop(['Air Quality'],axis=1)
y=data[['Air Quality']]
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
X_train
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
data['Air Quality'].unique()
```

```python
model=LogisticRegression(multi_class='ovr')
penalty = ['l1', 'l2', 'elasticnet']  # Regularization methods
C_values = [100, 10, 1.0, 0.1, 0.01]  # Regularization strength
solver = ['liblinear', 'saga']  # Only solvers compatible with l1 and elasticnet
multiclass=['ovr','multinomial']
param_grid = {
    'penalty': penalty,
    'C': C_values,
    'solver': solver
}
param_grid
```

```python
cv=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=param_grid,scoring='accuracy',cv=cv)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
y_pred
```

```python
print("The accuracy score for logistic regression with hyperparameter tuning is ",accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
kmeans=KNeighborsClassifier()
param_grid = {
    'n_neighbors': [2, 3, 4, 5, 6, 7, 8, 9, 10],  # Test a range of clusters
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],            # Algorithms for optimization
}
param_grid
```

```python
gridcv=GridSearchCV(estimator=kmeans,param_grid=param_grid,scoring='accuracy',cv=cv)
gridcv.fit(X_train,y_train)
```

```python
gridcv.best_params_
```

```python
y_pred=gridcv.predict(X_test)
y_pred
```

```python
print("The accuracy score for K means classifier  with hyperparamter tuning is ",accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
